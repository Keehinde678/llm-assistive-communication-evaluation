```python
import pandas as pd
# load the data
df=pd.read_csv("C:\\Users\\kehin\\Desktop\\Assistive Technology/outputs/model_outputs.csv")
# Quick checkoutputs/model_outputs.csv
print(f"Total rows: {len(df)}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Look at one prompt and all three model responses
row = df.iloc[0]

print(f"PROMPT ID: {row['prompt_id']}")
print(f"CATEGORY: {row['category']} | {row['disability_type']}")
print(f"\nPROMPT: {row['prompt_text'][:100]}...")
print(f"\nGPT-4 OUTPUT:\n{row['gpt4_output']}")
print(f"\nCLAUDE OUTPUT:\n{row['claude_output']}")
print(f"\nGROQ OUTPUT:\n{row['groq_output']}")
```

```python
# Create a rating template
rating_df = df[['prompt_id', 'category', 'disability_type', 
                'task_type', 'prompt_text', 
                'gpt4_output', 'claude_output', 'groq_output']].copy()

# Add rating columns for each model
for model in ['gpt4', 'claude', 'groq']:
    rating_df[f'{model}_clarity'] = ''
    rating_df[f'{model}_appropriateness'] = ''
    rating_df[f'{model}_accuracy'] = ''
    rating_df[f'{model}_consistency'] = ''
    rating_df[f'{model}_empathy'] = ''

# Save to Excel for raters
rating_df.to_excel("ratings/rating_sheet.xlsx", index=False)
print("Rating sheet saved to ratings/rating_sheet.xlsx")
```

```python
# Verify the rating sheet
rating_check = pd.read_excel("ratings/rating_sheet.xlsx")
print(f"Total rows: {len(rating_check)}")
print(f"\nColumns: {list(rating_check.columns)}")
# load and check the first rows
df=pd.read_excel("C:\\Users\\kehin\\Desktop\\Assistive Technology/ratings/rating_sheet.xlsx")
# Quick checkoutputs/model_outputs.csv
print(f"Total rows: {len(df)}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nFirst few rows:")
df.head()
```

```python
# View one prompt clearly for rating
def show_prompt(prompt_id):
    row = rating_check[rating_check['prompt_id'] == prompt_id].iloc[0]
    print(f"PROMPT ID: {row['prompt_id']}")
    print(f"CATEGORY: {row['category']} | {row['disability_type']}")
    print(f"TASK: {row['task_type']}")
    print(f"\nPROMPT:\n{row['prompt_text']}")
    print(f"\nGPT-4 OUTPUT:\n{row['gpt4_output']}")
    print(f"\nCLAUDE OUTPUT:\n{row['claude_output']}")
    print(f"\nGROQ OUTPUT:\n{row['groq_output']}")

show_prompt('P001')
```

```python
# Enter ratings for P001
ratings = {
    'P001': {
        'gpt4_clarity': 5, 'gpt4_appropriateness': 4, 'gpt4_accuracy': 4, 'gpt4_consistency': 5, 'gpt4_empathy': 4,
        'claude_clarity': 3, 'claude_appropriateness': 2, 'claude_accuracy': 3, 'claude_consistency': 3, 'claude_empathy': 3,
        'groq_clarity': 3, 'groq_appropriateness': 3, 'groq_accuracy': 4, 'groq_consistency': 3, 'groq_empathy': 4
    }
}

# Apply ratings to dataframe
for prompt_id, scores in ratings.items():
    for col, val in scores.items():
        rating_check.loc[rating_check['prompt_id'] == prompt_id, col] = val

# Verify
p001 = rating_check[rating_check['prompt_id'] == 'P001']
print("P001 Ratings entered:")
print(f"GPT-4:  Clarity={p001['gpt4_clarity'].values[0]}, Appropriateness={p001['gpt4_appropriateness'].values[0]}, Accuracy={p001['gpt4_accuracy'].values[0]}, Consistency={p001['gpt4_consistency'].values[0]}, Empathy={p001['gpt4_empathy'].values[0]}")
print(f"Claude: Clarity={p001['claude_clarity'].values[0]}, Appropriateness={p001['claude_appropriateness'].values[0]}, Accuracy={p001['claude_accuracy'].values[0]}, Consistency={p001['claude_consistency'].values[0]}, Empathy={p001['claude_empathy'].values[0]}")
print(f"Groq:   Clarity={p001['groq_clarity'].values[0]}, Appropriateness={p001['groq_appropriateness'].values[0]}, Accuracy={p001['groq_accuracy'].values[0]}, Consistency={p001['groq_consistency'].values[0]}, Empathy={p001['groq_empathy'].values[0]}")
# Save to Excel for manual rating
rating_check.to_excel("ratings/rating_sheet.xlsx", index=False)
print("Saved! Open ratings/rating_sheet.xlsx in Excel")
```

```python
# Kruskal-Wallis significance test
from scipy import stats

print("KRUSKAL-WALLIS TEST — STATISTICAL SIGNIFICANCE")
print("Tests whether differences between models are significant")
print("p < 0.05 = statistically significant difference\n")

dimensions = ['clarity', 'appropriateness', 'accuracy', 'consistency', 'empathy']

print(f"{'Dimension':<20} {'H-statistic':>12} {'p-value':>10} {'Significant':>12}")

for dim in dimensions:
    gpt4_scores   = rating_check[f'gpt4_{dim}'].dropna()
    claude_scores = rating_check[f'claude_{dim}'].dropna()
    groq_scores   = rating_check[f'groq_{dim}'].dropna()

    h_stat, p_value = stats.kruskal(gpt4_scores, claude_scores, groq_scores)
    significant = "Yes" if p_value < 0.05 else "No"
    print(f"{dim.capitalize():<20} {h_stat:>12.3f} {p_value:>10.4f} {significant:>12}")

# Overall
all_gpt4   = []
all_claude = []
all_groq   = []
for dim in dimensions:
    all_gpt4.extend(rating_check[f'gpt4_{dim}'].dropna().tolist())
    all_claude.extend(rating_check[f'claude_{dim}'].dropna().tolist())
    all_groq.extend(rating_check[f'groq_{dim}'].dropna().tolist())

h_stat, p_value = stats.kruskal(all_gpt4, all_claude, all_groq)
significant = "Yes" if p_value < 0.05 else "No"
print(f"{'OVERALL':<20} {h_stat:>12.3f} {p_value:>10.4f} {significant:>12}")
```

```python
# post-hoc test
from scipy.stats import mannwhitneyu
print("POST-HOC TEST — WHICH MODELS DIFFER?")
print("Mann-Whitney U test with Bonferroni correction")
print("p < 0.017 = significant (0.05/3 comparisons)\n")

comparisons = [
    ('gpt4', 'claude', 'GPT-4 vs Claude'),
    ('gpt4', 'groq',   'GPT-4 vs Groq'),
    ('claude', 'groq', 'Claude vs Groq')
]

dimensions = ['clarity', 'appropriateness', 'accuracy', 'consistency', 'empathy']

for dim in dimensions:
    print(f"\n{dim.upper()}")
    print(f"  {'Comparison':<20} {'U-stat':>10} {'p-value':>10} {'Significant':>12}")
    for m1, m2, label in comparisons:
        scores1 = rating_check[f'{m1}_{dim}'].dropna()
        scores2 = rating_check[f'{m2}_{dim}'].dropna()
        u_stat, p_value = mannwhitneyu(scores1, scores2, alternative='two-sided')
        significant = "Yes" if p_value < 0.017 else "No"
        print(f"  {label:<20} {u_stat:>10.1f} {p_value:>10.4f} {significant:>12}")
```

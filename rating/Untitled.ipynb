```python
import pandas as pd

df = pd.read_excel("ratings/rating_sheet.xlsx")
print(f"Loaded {len(df)} rows successfully")
df.head()
```

```python
# Create a clean sample of 20 responses to share
sample = pd.concat([
    df[df['disability_type'] == 'AAC'].head(7),
    df[df['disability_type'] == 'aphasia'].head(6),
    df[df['disability_type'] == 'intellectual_disability'].head(7)
])

# Keep only the columns needed for review
sample = sample[['prompt_id', 'disability_type', 'task_type',
                 'prompt_text', 'gpt4_output', 'claude_output', 'groq_output']]

# Rename columns to be reader friendly
sample.columns = ['Prompt ID', 'Disability Type', 'Task Type',
                  'Prompt', 'GPT-4 Response', 'Claude Response', 'Groq Response']

sample.to_excel("outputs/sample_responses_for_review.xlsx", index=False)
print(f"Sample file created — {len(sample)} responses")
print("Saved to: outputs/sample_responses_for_review.xlsx")
```
